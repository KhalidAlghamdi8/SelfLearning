```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()

import math 
median_Age = math.floor(df.Age.median())
median_Age
df.Age.fillna(median_Age)
```

```python
drop = df.drop(['PassengerId', 'Name', 'SibSp', 'Ticket', 'Cabin', 'Embarked'], axis = 'columns')
```

```python
drop
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()
drop['Sex_n'] = le_Sex.fit_transform(drop['Sex'])
```

```python
drop
```

```python
drop_final = drop.drop('Sex', axis='columns')
```

```python
drop_final
```

```python
inputs = drop_final.drop('Survived', axis='columns')
target = drop_final['Survived']
target
```

```python
inputs[['Age']] = inputs[['Age']].fillna(value=28)
```

```python

inputs
```

```python
inputs.isnull().sum().sum()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs,target)
```

```python
model.score(inputs,target)
```
